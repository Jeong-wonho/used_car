```python
urls=[]

with open('ss_tivoli-url.txt','r') as f:
    lines=f.readlines()
    for line in lines:
        urls.append(line.strip('\n'))
```

```python
#차량정보와 옵션을 크롤링합니다.
driver = webdriver.Chrome('chromedriver.exe')

try:
    # 에러나면 에러난 url 인덱스 번호와 에러를 프린트하고 멈춤
    for i in range(0,len(urls)):

        driver.get(urls[i])
        driver.refresh()
        time.sleep(0.5)
        if driver.find_elements_by_xpath('//*[@id="content"]/div[1]/div[2]/div[2]/div[1]/div[2]/div[1]/div/dl/dd/strong'):
            price=driver.find_elements_by_xpath('//*[@id="content"]/div[1]/div[2]/div[2]/div[1]/div[2]/div[1]/div/dl/dd/strong')[0].text
            price=price.replace('만원','0000')
            price=int(price.replace(',',''))
            가격.append(price)
        else:
            continue
        
        # 차량명 가져오기
        name=driver.find_elements_by_xpath('//*[@id="content"]/div[1]/div[2]/div[2]/div[1]/div[2]/div[1]/strong')[0].text
        car=name.split('\n')[0]
        
        # 브랜드와 모델명 나눠서 담기 (쉐보레 , 한국GM 겹쳐서 따로)
        if '쉐보레' in car:
            브랜드.append('한국GM 쉐보레')
            모델.append(car.split('한국GM 쉐보레')[1][1:])
        else:
            for brand in brandlist:
                if brand in car:
                    브랜드.append(brand)
                    모델.append(car.split(brand)[1][1:])
        
        
        # 차량정보 가져오기
        info=driver.find_elements_by_xpath('//*[@id="content"]/div[1]/div[2]/div[2]/div[5]/div')[0].text
        info.split('\n')
        
        #차량정보 리스트에 담기
        연식.append(info.split('\n')[4])
        주행거리.append(info.split('\n')[6])
        연료.append(info.split('\n')[8])
        변속기.append(info.split('\n')[10])
        연비.append(info.split('\n')[12])
        차종.append(info.split('\n')[14])
        배기량.append(info.split('\n')[16])
        색상.append(info.split('\n')[18])
        세금미납.append(info.split('\n')[20])
        압류.append(info.split('\n')[22])
        저당.append(info.split('\n')[24])
        보험사고정보.append(info.split('\n')[31])
        전손이력.append(info.split('\n')[33])
        침수이력.append(info.split('\n')[35])
        용도이력.append(info.split('\n')[37])
        소유자변경.append(info.split('\n')[39])
        연평균주행거리.append(info.split('\n')[-2][-9:-7])
        
        option=driver.find_elements_by_xpath('//*[@id="content"]/div[1]/div[2]/div[2]/div[6]/ul')[0].text
        
        # 주요옵션 1 , 0 으로
        for j in range(len(option_list)):
            if option_list[j] in option.split('\n')[0:-1]:
                options[j].append(1)
            else:
                options[j].append(0)
        
        # 사이드&커튼에어백 &때문에 리스트따로
        if '사이드&커튼에어백' in option.split('\n')[0:-1]:
            사이드n커튼에어백.append(1)
        else:
            사이드n커튼에어백.append(0)
        
    driver.quit()

except Exception as e:
    print(e)
    print(i)
```
