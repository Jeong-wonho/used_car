```python
#시각화에 필요한 라이브러리를 불러옵니다.
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import datetime
%matplotlib inline
```

```python
#matplotlib의 한글깨짐현상을 방지하기 위한 코딩
import matplotlib as mpl
import matplotlib.font_manager as fm
mpl.rcParams['axes.unicode_minus'] = False
path = 'C:/Windows/Fonts/malgun.ttf'
font_name = fm.FontProperties(fname=path, size=50).get_name()
plt.rc('font', family=font_name)
```

```python
df = pd.read_csv('오늘이발표날.csv', index_col = 0)
```

```python
#데이터 전처리후 브랜드별 중고차 수량
#
sns.countplot(df.브랜드 , alpha = 0.5, order = df['브랜드'].value_counts().index)
plt.grid('white')
```

```python
#데이터 전처리후 모델별 중고차 수량
sns.countplot(df.모델 , alpha = 0.5, order = df.모델.value_counts().index)
plt.xticks(rotation = - 45 )
plt.grid('white')
```

```python
#데이터 전처리후 옵션별 중요도 파악
figure, ((ax1,ax2,ax3), (ax4,ax5,ax6), (ax7,ax8,ax9)) = plt.subplots(nrows=3, ncols=3,figsize=(7,5))
figure.set_size_inches(18,12)

axs = [ax1,ax2,ax3,ax4,ax5,ax6,ax7,ax8,ax9]
options = ['썬루프', '내비게이션', '오토라이트', '주차감지센서', '가죽시트', '열선시트', '전동접이사이드미러', '크루즈컨트롤', '알루미늄휠']

for a,o in zip(axs,options):
    sns.countplot(data=df, x=o, ax=a)
plt.show()
```

```python
#데이터 전처리후 옵션별 중요도 파악하기
figure, ((ax1,ax2,ax3), (ax4,ax5,ax6), (ax7,ax8,ax9)) = plt.subplots(nrows=3, ncols=3,figsize=(7,5))
figure.set_size_inches(18,12)

axs = [ax1,ax2,ax3,ax4,ax5,ax6,ax7,ax8,ax9]
options = ['스마트키', '통풍시트', '하이패스룸밀러', '블루투스', '타이어공기압장치', '레인선서와이퍼', '핸들열선', '차선이탈경보', '사이드n커튼에어백']

for a,o in zip(axs,options):
    sns.countplot(data=df, x=o, ax=a)
plt.show()
```

```python
#데이터 전처리후 너무 적게 있던 연식들을 삭제하였습니다.
sns.countplot(df['연식'], order = df.연식.value_counts().index)
```

```python
# 데이터 전처리후 연식별 옵션 갯수
figure, ((ax1,ax2,ax3), (ax4,ax5,ax6), (ax7, ax8, ax9)) = plt.subplots(nrows=3, ncols=3,figsize=(7,5))
figure.set_size_inches(18,12)

axs = [ax1,ax2,ax3,ax4,ax5,ax6,ax7,ax8,ax9]
options = ['썬루프', '내비게이션', '오토라이트', '주차감지센서', '가죽시트', '열선시트', '전동접이사이드미러', '크루즈컨트롤', '알루미늄휠']

for a,o in zip(axs,options):
    sns.countplot(data=df, x ='연식' , hue = o, alpha = 0.7, ax=a,order = df['연식'].value_counts().index)
    a.legend(loc = 'upper right',title = o)
plt.show()
```
