## machine learning 돌리기 전 라벨인코더를 통한 분류

```python
import pandas as pd
import numpy as np
```

```python
#라벨인코더를 통한 분류 및 숫자로 변경
from sklearn import preprocessing
le = preprocessing.LabelEncoder()
le.fit(sample['차종'])
le.classes_
sample['차종'] = le.transform(sample['차종'])
```
